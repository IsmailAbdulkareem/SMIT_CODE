# **Q7:** `while loop` ka use karke user se tab tak numbers input lein jab tak wo `0` enter na karde. End mein saare numbers ka total dikhayein.

````python
total = 0

while True:
    num = int(input("Number enter karein (stop karne ke liye 0): "))

    if num == 0:
        break

    total += num

print("Saare numbers ka total:", total)
````

# **Q8:** Ek list `numbers = [12, 45, 7, 23, 56, 90, 11]` mein se sabse bada (Maximum) number find karein loop ke zariye.

````python
# type your code here
numbers = [12, 45, 7, 23, 56, 90, 11]

max_num = numbers[0]

for num in numbers:
    if num > max_num:
        max_num = num

print("Sabse bada number:", max_num)
````

# **Q9:** User se ek string input lein (e.g., 'Python') aur loop ka use karke us string ko reverse print karein.

# **Q10:** 1 se 50 tak ke numbers check karein. Agar number 3 se divide hota hai to 'Fizz' print karein, agar 5 se hota hai to 'Buzz' print karein, aur agar dono se hota hai to 'FizzBuzz' print karein.

````python
# type your code here
for num in range(1, 51):

    if num % 3 == 0 and num % 5 == 0:
        print("FizzBuzz")

    elif num % 3 == 0:
        print("Fizz")

    elif num % 5 == 0:
        print("Buzz")

    else:
        print(num)
````

# **Q11:** Ek list di gayi hai `items = [1, 2, 3, 4, 5]`. Loop ka use karke ek nayi list banayein jis mein har number ka **Square** (e.g., 1, 4, 9...) ho.

````python
# type your code here
items = [1, 2, 3, 4, 5]

squares = []

for num in items:
    squares.append(num ** 2)

print("Square wali list:", squares)
````

# **Q12:** User se ek number lein aur check karein ke wo **Prime Number** hai ya nahi (using loops).

````python
# type your code here
num = int(input("Ek number enter karein: "))

is_prime = True

if num <= 1:
    is_prime = False
else:
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

if is_prime:
    print(num, "Prime Number hai")
else:
    print(num, "Prime Number nahi hai")
````

# **Q16:** User se ek number input lein aur us ka **Factorial** find karein (e.g., 5! = 5*4*3*2*1).

````python
# type your code here
num = int(input("Ek number enter karein: "))

factorial = 1

for i in range(1, num + 1):
    factorial *= i

print("Factorial =", factorial)
````

# **Q18:** Ek manual counting karein ke ek string (e.g., 'Learning Python is fun') mein kitne **Vowels** (a, e, i, o, u) hain.

````python
# type your code here
text = input("Ek string enter karein: ")

count = 0

for char in text.lower():
    if char in "aeiou":
        count += 1

print("Total vowels:", count)
````

# **Q19:** **Fibonacci Sequence** ke pehle 10 numbers print karein (0, 1, 1, 2, 3, 5, 8...)

````python
# type your code here
a = 0
b = 1

print("Fibonacci Sequence:")

for i in range(10):
    print(a, end=" ")

    c = a + b
    a = b
    b = c
````

# **Q20:** Ek 'Number Guessing Game' banayein jahan program ek number (1-20) fix kare aur user ko 3 chances de guess karne ke liye. `while loop` aur `break` ka use karein.

````python
# type your code here
secret_number = 7
chances = 3

while chances > 0:

    guess = int(input("1 se 20 tak number guess karein: "))

    if guess == secret_number:
        print("Congratulations! Aap ne sahi number guess kar liya.")
        break

    else:
        chances -= 1
        print("Wrong guess! Remaining chances:", chances)

if chances == 0:
    print("Game Over! Correct number tha:", secret_number)
````
