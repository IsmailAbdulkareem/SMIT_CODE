### 43. Ek simple calculator banayein (+, -, *, /) symbols ke basis par operation perform karein.

````python
# Write your code here
num1 = float(input("Pehla number enter karein: "))
operator = input("Operator enter karein (+, -, *, /): ")
num2 = float(input("Doosra number enter karein: "))

if operator == "+":
    result = num1 + num2
elif operator == "-":
    result = num1 - num2
elif operator == "*":
    result = num1 * num2
elif operator == "/":
    if num2 != 0:
        result = num1 / num2
    else:
        print("Error: Zero se divide nahi kar sakte!")
        exit()
else:
    print("Invalid operator!")
    exit()

print(f"{num1} {operator} {num2} = {result}")
````

### 44. User ka weight aur height le kar BMI category batayein (Underweight, Normal, Overweight).

````python
# Write your code here
weight = float(input("Weight (kg) mein enter karein: "))
height = float(input("Height (meters) mein enter karein: "))

bmi = weight / (height ** 2)

if bmi < 18.5:
    category = "Underweight"
elif bmi < 25:
    category = "Normal"
elif bmi < 30:
    category = "Overweight"
else:
    category = "Obese"

print(f"Weight: {weight} kg, Height: {height} m")
print(f"BMI: {bmi:.2f}")
print(f"Category: {category}")
````

### 45. ATM Machine Logic: Pehle PIN check karein, agar sahi ho toh Amount poochiye, agar account balance se kam amount ho toh 'Withdrawal Success'.

````python
# Write your code here
PIN = "123"
acc_balance = 9000
ac_pin = input("Enter your PIN: ")

if ac_pin == PIN:
    amount = int(input("Enter withdrawal amount: "))
    if amount <= acc_balance:
        acc_balance -= amount
        print("Withdrawal successful!")
        print(f"Remaining balance: {acc_balance}")
    else:
        print("Failed to proceed, insufficient balance")
else:
    print("Wrong PIN!")
````

### 58. Electricity Bill: Pehle 100 units free, next 100 units Rs 5/unit, uske baad Rs 10/unit.

### 60. Final Challenge: Ek mini-project logic banayein jisme user se Username, Password, aur ek Security Question poochein nested loops aur if-else ke saath.
