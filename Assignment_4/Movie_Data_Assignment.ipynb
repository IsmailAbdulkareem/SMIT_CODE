### **Question 1**
Dataset (`movie.csv`) ko pandas mein load karein aur uske pehle 5 rows display karein.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('movie.csv')
df['Released_Year'] = pd.to_numeric(df['Released_Year'], errors='coerce')
df['Runtime_Min'] = df['Runtime'].str.replace(' min', '').astype(int)
df['Gross'] = df['Gross'].fillna(df['Gross'].median())
df.head()
```

### **Question 21**

### **Question 40**
(Visualization) Top 5 highest rated movies aur unke 'Meta_score' vs 'IMDB_Rating' (scaled to 100) ka ek Side-by-Side Comparison Bar Chart banayein.

```python
top5_rated = df.nlargest(5, 'IMDB_Rating')[['Series_Title', 'IMDB_Rating', 'Meta_score']].copy()
top5_rated['Meta_x10'] = top5_rated['Meta_score'] * 10
x = np.arange(len(top5_rated))
width = 0.35
plt.figure(figsize=(12, 6))
plt.bar(x - width/2, top5_rated['IMDB_Rating'] * 10, width, label='IMDB Rating (scaled)', color='navy')
plt.bar(x + width/2, top5_rated['Meta_x10'], width, label='Meta Score', color='crimson')
plt.xticks(x, top5_rated['Series_Title'], rotation=45, ha='right')
plt.xlabel('Movie')
plt.ylabel('Score')
plt.title('Top 5 Highest Rated Movies: IMDB vs Meta Score')
plt.legend()
plt.tight_layout()
plt.show()
```
