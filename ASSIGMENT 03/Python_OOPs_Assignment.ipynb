**Q1:** Aik `Student` class banayein jismein `name` aur `roll_number` private variables (`__`) hon. In variables ko access aur modify karne ke liye getter aur setter methods likhein.

```python
## Write your code here for Q1
class Student:
    def __init__(self, name, roll_number):
        self.__name = name              # private variable
        self.__roll_number = roll_number

# Getters
    def get_name(self):
        return self.__name
    def get_roll_number(self):
        return self.__roll_number

# Setters

    def set_name(self, name):
        self.__name = name

    def set_roll_number(self , roll_number):
        self.__roll_number = roll_number

s =Student ("Ismail" , 143)
print(s.get_name())
print(s.get_roll_number())

s.set_name("Ali")
s.set_roll_number(111)

print(s.get_name())
print(s.get_roll_number())
```

**Q2:** Aik `BankAccount` class banayein jismein `__balance` private variable ho. Aik `deposit()` aur aik `withdraw()` method banayein, check karein ke withdraw karte waqt balance zero se kam na ho.

```python
#Write your code here for Q2
class BankAccount:
    def __init__(self, balance=0):
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount
        print(f"Deposited: {amount} | Balance: {self.__balance}")

    def withdraw(self, amount):
        if amount > self.__balance:
            print("Insufficient balance!")
        else:
            self.__balance -= amount
            print(f"Withdrawn: {amount} | Balance: {self.__balance}")

    def get_balance(self):
        return self.__balance

acc = BankAccount(1000)
acc.deposit(500)
acc.withdraw(200)
acc.withdraw(2000)   # Insufficient balance!
```

**Q4:** Protected variable (`_`) aur Private variable (`__`) ke beech kya farq hota hai? Aik simple code ke zariye dono ka use-case dikhayein.

```python
# Write your code here for Q4
class MyClass:
    def __init__(self):
        self._protected = "I am protected"   #This is  single underscore = convention only
        self.__private = "I am private"      #This is  double underscore = name mangling

    def show(self):
        print(self._protected)
        print(self.__private)

obj = MyClass()

# Protected — accessible (but shouldn't be, by convention)
print(obj._protected)       # works

# Private — NOT directly accessible
# print(obj.__private)      

# Private access sirf iss tarah possible (name mangling):
print(obj._MyClass__private)  # works but avoid this
```

**Q9:** Child class ke constructor (`__init__`) ke andar parent class ke constructor ko call karne ke liye `super()` ka use kaise hota hai? Aik practical example likhein.

```python
# Write your code here for Q9
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        print(f"Person created: {self.name}")

class Teacher(Person):
    def __init__(self, name, age, subject):
        super().__init__(name, age)   # parent ka __init__ call
        self.subject = subject
        print(f"Teacher of: {self.subject}")

t = Teacher("Sir Ahmed", 35, "Python")
print(t.name, t.age, t.subject)
```

**Q10:** Hierarchical Inheritance kya hoti hai? Aik parent class `Shape` banayein aur us se do alag child classes `Circle` aur `Square` banakar samjhayein.

```python
# Write your code here for Q10
class Shape:
    def area(self):
        print("Calculating area...")

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        print(f"Circle Area: {3.14 * self.radius ** 2}")

class Square(Shape):
    def __init__(self, side):
        self.side = side

    def area(self):
        print(f"Square Area: {self.side ** 2}")

c = Circle(5)
s = Square(4)
c.area()   # Circle Area: 78.5
s.area()   # Square Area: 16
```

**Q11:** Method Overriding kya hai? Parent class `Bird` banayein jisme `fly()` method ho jo print kare 'Most birds can fly'. Child class `Ostrich` banayein jo is method ko override kare aur print kare 'Ostriches cannot fly'.

**Q13:** Operator Overloading kya hoti hai? Do objects ko `+` operator ke zariye add karne ke liye `__add__` magic method ka use kaise karte hain? Code likhein.

```python
# Write your code here for Q13
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

v1 = Vector(1, 2)
v2 = Vector(3, 4)
v3 = v1 + v2   # __add__ automatically call hoga
print(v3)
```

**Q17:** Q16 wali `Computer` class ko inherit karke do child classes `Laptop` aur `Desktop` banayein aur dono mein `process()` method ko mandatory implement karein.

```python
# Write your code here for Q17
from abc import ABC, abstractmethod

class Computer(ABC):
    @abstractmethod
    def process(self):
        pass

class Laptop(Computer):
    def process(self):
        print("Laptop: Processing on battery")

class Desktop(Computer):
    def process(self):
        print("Desktop: Processing on power supply")

l = Laptop()
d = Desktop()
l.process()
d.process()
```

**Q18:** Aik abstract class `Car` banayein jisme aik concrete (normal) method `start_engine()` ho aur aik abstract method `drive()` ho. Child class bana kar dono ko test karein.

```python
# Write your code here for Q18
from abc import ABC, abstractmethod

class Car(ABC):
    def start_engine(self):          # concrete method — har car ke liye same
        print("Engine started!")

    @abstractmethod
    def drive(self):                 # har car apne tarah drive karti hai
        pass

class ElectricCar(Car):
    def drive(self):
        print("Electric car drives silently")

ec = ElectricCar()
ec.start_engine()   # parent se
ec.drive()          # child ne implement kiya
```

**Q20:** Aik real-world scenario (jaise Payment System) design karein jahan `Payment` aik abstract class ho aur `CreditCardPayment` aur `UPIPayment` uski subclasses hon.

```python
# Write your code here for Q20
from abc import ABC, abstractmethod

class Payment(ABC):
    def __init__(self, amount):
        self.amount = amount

    @abstractmethod
    def pay(self):
        pass

    def receipt(self):
        print(f"Payment of Rs.{self.amount} processed.")

class CreditCardPayment(Payment):
    def pay(self):
        print(f"Paying Rs.{self.amount} via Credit Card")
        self.receipt()

class UPIPayment(Payment):
    def pay(self):
        print(f"Paying Rs.{self.amount} via UPI / EasyPaisa")
        self.receipt()

# Usage
p1 = CreditCardPayment(5000)
p2 = UPIPayment(1500)

p1.pay()
p2.pay()
```
